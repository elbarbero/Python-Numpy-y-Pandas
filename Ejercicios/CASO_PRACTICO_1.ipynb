```python
import pandas as pd
import numpy as np
import requests
import io
```

```python
def GetTablesInUrl(url):
    try:
        url = url
        r = requests.get(url)
        return pd.read_html(io.StringIO(str(r.text)))
    except ValueError as ex:
        print(f"No se han encontrado tablas:\n {ex}")
        return None

df_lista_pob = GetTablesInUrl("https://es.wikipedia.org/wiki/Anexo:Pa%C3%ADses_y_territorios_dependientes_por_poblaci%C3%B3n")
df_lista_pib = GetTablesInUrl("https://www.worldometers.info/gdp/gdp-by-country/")
df_lista_rpc = GetTablesInUrl("https://en.wikipedia.org/wiki/List_of_countries_by_average_wage")
```

```python
def RemoveNANRows(dataframe, name):
    original_rows = dataframe.shape[0]
    dataframe.dropna(inplace=True)
    wrong_rows = original_rows-dataframe.shape[0]
    print(f"- En el dataframe de {name} hay {wrong_rows} filas erroneas")

RemoveNANRows(df_ventas, "ventas")
RemoveNANRows(df_clientes, "clientas")
RemoveNANRows(df_productos, "productos")
RemoveNANRows(df_pob, "población")
RemoveNANRows(df_pib, "PIB")
RemoveNANRows(df_rpc, "renta per capita")
```
