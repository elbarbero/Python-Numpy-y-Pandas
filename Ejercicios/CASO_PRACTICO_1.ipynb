```python
import pandas as pd
import numpy as np
import requests
import io
```

```python
def GetTablesInUrl(url):
    try:
        url = url
        r = requests.get(url)
        return pd.read_html(io.StringIO(str(r.text)))
    except ValueError as ex:
        print(f"No se han encontrado tablas:\n {ex}")
        return None

df_lista_pob = GetTablesInUrl("https://es.wikipedia.org/wiki/Anexo:Pa%C3%ADses_y_territorios_dependientes_por_poblaci%C3%B3n")
df_lista_pib = GetTablesInUrl("https://www.worldometers.info/gdp/gdp-by-country/")
df_lista_rpc = GetTablesInUrl("https://en.wikipedia.org/wiki/List_of_countries_by_average_wage")
```

<h1>PREGUNTA 12</h1>
<p>¿En qué año se han obtenido un mayor número de ventas?
</p>

```python
df_date = pd.DataFrame(df_global.groupby("Fecha compra")["Cantidad"].agg("max")).sort_values(by=['Cantidad'], ascending=False)
df_date
```

```python
#np.datetime64(df_date.index.values[0]).year
datetimes = pd.to_datetime(df_date.index)
datetimes.year
df_date["Día"] = pd.to_datetime(df_date.index).day
df_date["Mes"] = pd.to_datetime(df_date.index).month
df_date["Año"] = pd.to_datetime(df_date.index).year
```

```python
df_date
```

```python
df_date_sum = pd.DataFrame(df_date.groupby("Año")["Cantidad"].agg("sum")).sort_values(by=['Cantidad'], ascending=False)
df_date_sum
```

```python
cantidad = df_date_sum.loc[2021, "Cantidad"]
print(f"El año que más se vendió fue el año {df_date_sum.index.values[0]} con {cantidad} ventas.")
```
