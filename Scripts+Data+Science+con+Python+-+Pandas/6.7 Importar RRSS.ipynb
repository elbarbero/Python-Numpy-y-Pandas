# Guardar tweets del timeline de nuestro usuario

```python
def get_tweets(self, since_id):

        since_id = str(int(since_id) - 1)

        tweets = []
        for status in tweepy.Cursor(self.user_timeline,
                             user_id=api.me(),
                             since_id=since_id).items():

            tweets.append(status._json)

        return tweets 
```

```python
json_data = get_tweets(api,100)
```
